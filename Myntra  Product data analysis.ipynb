```python
# Here we can choose only the relevant column's data  for analysis
key_columns=['BrandName', 'Category', 'Individual_category', 'category_by_Gender', 
               'DiscountPrice (in Rs)', 'OriginalPrice (in Rs)', 'DiscountOffer', 
               'Ratings', 'Reviews', 'SizeOption']
for column in key_columns:
    print(f"\nUnique values in {column}:")
    print(data[column].unique())
```

```python
#lets check for the missing values in the Key Columns
missing_values = data[key_columns].isnull().sum()
print("Missing Values in Key Columns:")
print(missing_values)
```

```python
missing_data=data.isnull().sum()
missing_value_percentage=(missing_data/len(data))*100
missing_data_summary=pd.DataFrame({
    "Missing Value": missing_data,
    "Percentage":missing_value_percentage
})
print(missing_data_summary)
```

```python
#As discount can be an important column so we replace the 'DiscountPrice (in Rs)' column with the median
median_discount_price= data["DiscountPrice (in Rs)"].median()
data["DiscountPrice (in Rs)"].fillna(median_discount_price, inplace=True)

# Replacing the discount offer with the mod value
mod_discountoffer=data["DiscountOffer"].mode()[0]
data["DiscountOffer"].fillna(mod_discountoffer,inplace=True)

missing_values_after=data.isnull().sum()
missing_values_after[['DiscountPrice (in Rs)', 'DiscountOffer']]
print("Missing Value now:",data.isnull().sum())
```

```python
# Replacing the missing values in  Ratings column with the mod
mod_ratings=data["Ratings"].mode()[0]
data["Ratings"].fillna(mod_ratings,inplace=True)

# Handling missing values in review by replacing them 0
data["Reviews"] = data["Reviews"].fillna(0)
missing_values_after=data.isnull().sum()
print("Missing Value now:",data.isnull().sum())
```

# Visualization 


```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')
fig,axes= plt.subplots(2,2, figsize=(15, 12))

# Plotting DiscountPrice
sns.histplot(data["DiscountPrice (in Rs)"], kde=True, ax=axes[0, 0], color='blue')
axes[0, 0].set_title('Distribution of Discount Price')

# Plotting OriginalPrice
sns.histplot(data["OriginalPrice (in Rs)"], kde=True, ax=axes[0, 1], color='green')
axes[0, 1].set_title('Distribution of Original Price')

# Plotting Ratings
sns.histplot(data["Ratings"], kde=True, ax=axes[1, 0], color='red')
axes[1, 0].set_title('Distribution of Ratings')

# Plotting Reviews
sns.histplot(data["Reviews"], kde=True, ax=axes[1, 1], color='purple')
axes[1, 1].set_title('Distribution of Reviews')

# Show the plots
plt.tight_layout()
plt.show()
```

# Correlation matrix

```python

corr = data[["DiscountPrice (in Rs)", "OriginalPrice (in Rs)", "Ratings", "Reviews"]].corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=1, linecolor='black')
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```
